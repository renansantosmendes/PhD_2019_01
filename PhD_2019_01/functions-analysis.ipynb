```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
algorithm = 'ONMOEAD'
date = '2020_3_21'
instance_name = 'r050n12tw10k11'
file_path = '/home/renansantos/Área de Trabalho/Doutorado/PhD_2019_01/PhD_2019_01/Results_2020'
file_name = 'onmoead-initial_population.csv'
```

```python
f = open(os.path.join(file_path, algorithm, instance_name + '_' + date, file_name),'r',encoding='utf-8')
data = f.readlines()
data[0]
splitted_data = [i.split('\t')[1] for i in data]
splitted_data = [[float(j) for j in i.split(',')] for i in splitted_data]
```

```python
data_df = pd.DataFrame(data=splitted_data, columns = 'f1 f2 f3 f4 f5 f6 f7 f8'.split())
```
