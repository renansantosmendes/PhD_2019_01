```python
import os
import numpy as np
import pandas as pd
import matplotlib. pyplot as plt
from pygmo import *
from mpl_toolkits.mplot3d import Axes3D
import methods_updated as mt
%matplotlib inline
```

# Instância r050n12tw10k4

## MOEAD-PCA

```python
file = 'moead-original-pareto-execution-'
instance = 'r050n12tw10k4'
algorithm_folder_1 = 'MOEAD_PCA'
algorithm_name_1 = 'MOEAD-PCA'
hv_moead_pca_r050, hv_moead_pca_r050_df = mt.plot_hypervolume_convergence(
    os.path.join(algorithm_folder_1, instance),
    file,
    8,
    plot=False)
```

## MOEAD

```python
file = 'moead-original-pareto-execution-'
instance = 'r050n12tw10k4'
algorithm_folder_2 = 'MOEAD'
algorithm_name_2 = 'MOEAD'
hv_moead_r050, hv_moead_r050_df = mt.plot_hypervolume_convergence(
    os.path.join(algorithm_folder_2, instance),
    file,
    8,
    plot=False)
```

```python
hv_moead_pca_r050[0] = 1.6081469550308707e+35

plt.figure(figsize=(10,8))
plt.plot(hv_moead_pca_r050, label=algorithm_name_1)
plt.plot(hv_moead_r050, label=algorithm_name_2)
plt.xlabel('Evaluation')
plt.legend()
plt.show()
```

```python
data_to_plot = [hv_moead_pca_r050_df.iloc[299:].values.tolist()[0],
                hv_moead_r050_df.iloc[299:].values.tolist()[0]
               ]
plt.figure(figsize=(10, 7))
plt.boxplot(data_to_plot, labels=[algorithm_name_1,
                                  algorithm_name_2
                                  ])
plt.show()
```

```python
mt.random_test(hv_moead_pca_r050_df.iloc[299:], hv_moead_r050_df.iloc[299:], 10, False)
```
