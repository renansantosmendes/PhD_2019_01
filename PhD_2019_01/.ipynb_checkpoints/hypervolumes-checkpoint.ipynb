```python
import os
import pandas as pd
from pygmo import *
```

```python
algorithm = 'ONMOEAD'
date = '2020_3_21'
instance_name = 'r050n12tw10k11'
file_path = '/home/renansantos/Área de Trabalho/Doutorado/PhD_2019_01/PhD_2019_01/Results_2020'
file_name = 'onmoead-reduced-pareto-execution-0.txt'
```

```python
f = open(os.path.join(file_path, algorithm, instance_name + '_' + date, file_name),'r',encoding='utf-8')
data = f.readlines()
splitted_data = ' '.join([i for i in data]).split('#\n')
splitted_data = [i for i in splitted_data if len(i) > 1]
```

```python
hv_pareto = []
indexes = []
for k in range(len(splitted_data)):
    converted = [i.strip().split(',') for i in splitted_data[k].split('\n') if len(i) > 1]
    pop = [[float(j) for j in i] for i in converted]
    hv = hypervolume(pop)
    hv_pareto.append(hv.compute([100000,100000,100000]))    
    indexes.append(100*k)
```

```python
df1 = pd.DataFrame(data=[indexes, hv_pareto]).T
df1.columns = 'Evaluation HV'.split()
```

```python
df1.plot(x='Evaluation', y='HV',figsize=(10,8))
```

```python
algorithm = 'ONMOEAD'
date = '2020_3_21'
instance_name = 'r050n12tw10k11'
file_path = '/home/renansantos/Área de Trabalho/Doutorado/PhD_2019_01/PhD_2019_01/Results_2020'
file_name = 'onmoead-original-pareto-execution-0.txt'
```

```python
f = open(os.path.join(file_path, algorithm, instance_name + '_' + date, file_name),'r',encoding='utf-8')
data = f.readlines()
splitted_data = ' '.join([i for i in data]).split('#\n')
splitted_data = [i.replace('\t','') for i in splitted_data if len(i) > 1]
```

```python
hv_pareto = []
indexes = []
for k in range(len(splitted_data)):
    converted = [i.strip().split(',') for i in splitted_data[k].split('\n') if len(i) > 1]
    pop = [[float(j) for j in i] for i in converted]
    hv = hypervolume(pop)
    hv_pareto.append(hv.compute([600000,1000,1000,1000,10000,1000,10000, 1]))    
    indexes.append(100*k)
```

```python
df2 = pd.DataFrame(data=[indexes, hv_pareto]).T
df2.columns = 'Evaluation HV'.split()
```

```python
df2.plot(x='Evaluation', y='HV',figsize=(10,8))
```
